```python
# 第一步：声明Notebook的类型
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
# 第二布导入所需模块
from pyecharts.charts import Line
from pyecharts import options as opts # 在pyecharts中，一切皆options
from pyecharts.globals import ThemeType # ptecharts内置了十余种图表主题
```

```python
line = (
    Line(init_opts = opts.InitOpts(height = "400px",width = "800px",theme=ThemeType.DARK))
    .add_xaxis([str(i) for i in range(2015,2026)])
    .add_yaxis("温度变化趋势",
               [17.6,18.9,21.3,20.2,22.3,23.4,24.0,23.2,20.1,20.5],
               is_smooth = True
              )
    .set_global_opts(
        yaxis_opts = opts.AxisOpts(name = "温度（℃）"),
        title_opts = opts.TitleOpts(title = "温度变化"),
        toolbox_opts = opts.ToolboxOpts(),
        datazoom_opts = opts.DataZoomOpts()
    )
    .set_series_opts(
        label_opts = opts.LabelOpts(position = "top")
    )
)
line.render_notebook()
```
