```python
# 声明notebook类型
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
# 导入模块
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.globals import ThemeType
```

```python
bar = (
    Bar(init_opts = opts.InitOpts(width = "800px",height = "400px",theme = ThemeType.LIGHT))
    .add_xaxis(["王者荣耀","和平精英","蛋仔派对","英雄联盟"])
    .add_yaxis("青少年",[310,456,535,462])
    .add_yaxis("中老年",[260,305,402,301])
    .set_global_opts(
        title_opts = opts.TitleOpts(title = "各类型游戏玩家对比"), # 设置标题
        toolbox_opts = opts.ToolboxOpts(), # 设置工具箱
        datazoom_opts = opts.DataZoomOpts() # 设置数据缩放
                    )
    .set_series_opts(
        label_opts = opts.LabelOpts(position = "top") # 设置标签位置
                    )
    
     )
```

```python
bar.render_notebook()
```

```python
# 导入所需模块
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB  # 声明notebook类型
from pyecharts.charts import Bar #导入柱状图模块
from pyecharts import options as opts 
from pyecharts.globals import ThemeType # 导入图表主题模块
```

```python
bar = (
    Bar(init_opts = opts.InitOpts(width = "800px",height = "400px",theme = ThemeType.LIGHT))
    .add_xaxis(["李白","杜甫","王维"])
    .add_yaxis("男生",[100,123,96])
    .add_yaxis("女生",[152,126,87])
    .set_global_opts(
        title_opts=opts.TitleOpts(title="各诗人粉丝数量"),
        toolbox_opts=opts.ToolboxOpts(),
        datazoom_opts=opts.DataZoomOpts()
    )
    .set_series_opts(
        label_opts=opts.LabelOpts(position="top")
    )
)
```

```python
bar.render_notebook()
```
