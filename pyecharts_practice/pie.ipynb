```python
# 声明Notebook的类型
from pyecharts.globals import CurrentConfig,NotebookType
CurrentConfig.NOTEBOOK_TYPE = NotebookType.JUPYTER_LAB
```

```python
# 导入相关模块
from pyecharts.charts import Pie # 导入饼状图
from pyecharts import options as opts # 一切皆options
from pyecharts.globals import ThemeType # 选择内置主题使用
```

```python
pie = (
    Pie()
    .add("",
         [("外卖",44),("生活用品",54),("餐饮",89),("娱乐",37)],
         radius = ["30%","75%"],
         rosetype = "radius"
    )
    .set_global_opts(
        title_opts = opts.TitleOpts(title = "个人支出情况")
    )
    .set_series_opts(
        label_opts = opts.LabelOpts(formatter = "{b}:{d}%")
    )
)
pie.render_notebook()
```
